### Start changing this to an appropriately configured dataset with all trials as rows and columns just holding values

```python
rot_df = pd.DataFrame(train_data.unstack().reset_index()[:])
rot_df.rename(columns={'level_0':'location','Date':'date',0:'total_solar'}, inplace=True)
#, columns=['location','date','total_solar'])
```

```python
rot_df.set_index('date',inplace=True)
```

```python
#rot_df
```

### Now merge with the location information

```python
loc_info = pd.read_csv('../data/kaggle_solar/station_info.csv', index_col='stid')
```

```python
temp_df = rot_df.reset_index('location').merge(right=loc_info, left_on='location', right_index=True)
temp_df
temp_df = temp_df.reset_index()
temp_df.set_index(['date','location'],inplace=True)
```
